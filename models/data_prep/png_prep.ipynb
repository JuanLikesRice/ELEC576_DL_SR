```python
import os
from PIL import Image
import shutil
def count_png_files(directory):
    png_files = [f for f in os.listdir(directory) if f.lower().endswith(".png")]
    print(f"Number of PNG files in {directory}: {len(png_files)}")
    return len(png_files)

print(f"All top largest images have been combined into: {output_dir}")
dir_i = r"C:\Users\Turog\OneDrive\Documents\GitHub\576_DL_SuperRes\data\DIV2K"
dir_j = r"C:\Users\Turog\OneDrive\Documents\GitHub\576_DL_SuperRes\data\ssid"
dir_k = r"C:\Users\Turog\OneDrive\Documents\GitHub\576_DL_SuperRes\data\waves"
count_png_files(dir_i)
count_png_files(dir_j)
count_png_files(dir_k)


output_dir = r"C:\Users\Turog\OneDrive\Documents\GitHub\576_DL_SuperRes\data\combined_largest_images_test_model"
os.makedirs(output_dir, exist_ok=True)






top_i =  500 
top_j =  500  
top_k =  600 

def get_top_n_largest_images(directory, n):
    images = []
    for filename in os.listdir(directory):
        if filename.lower().endswith(".png"):
            img_path = os.path.join(directory, filename)
            with Image.open(img_path) as img:
                width, height = img.size
                area = width * height
                images.append((area, filename, img_path))
    
    # Sort images by area (largest first) and return the top N
    images.sort(reverse=True, key=lambda x: x[0])
    return images[:n]

top_images_i = get_top_n_largest_images(dir_i, top_i)
top_images_j = get_top_n_largest_images(dir_j, top_j)
top_images_k = get_top_n_largest_images(dir_k, top_k)

# Combine all selected images and copy to the output directory
all_top_images = top_images_i + top_images_j + top_images_k

for _, filename, img_path in all_top_images:
    shutil.copy(img_path, os.path.join(output_dir, filename))
    print(f"Copied: {img_path} -> {output_dir}")

print(f"All top largest images have been combined into: {output_dir}")
```
